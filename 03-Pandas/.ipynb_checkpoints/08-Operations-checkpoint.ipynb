# 08 - Operations

```python
##In this lecture we are going to review some of the operations with Pandas and learn about some other useful 
##operations. 

import numpy as np
import pandas as pd 
```

```python
df = pd.DataFrame({'col1':[1,2,3,4], 
                   'col2':[444,555,666,444], 
                   'col3':['abc', 'def', 'ghi', 'xyz']})

df.head() ##Outputs the specified number of rows from the calling Data Frame - Pass an int as argument 
```

```python
### Finding unique values - Three useful methods 

## 1 - .unique() -> Returns all the unique values in a specified column as a np.array
df['col2'].unique()
```

```python
## We obtain the number of unique of values calling the len() method of the above result
len(df['col2'].unique())
```

```python
## 2 - .nunique()
## We use the nunique() method to do the above in a single step 

df['col2'].nunique()
```

```python
## 3 - .value_counts() -> Returns a table of the occurence of unique values as a table, e.g. 

df['col2'].value_counts()
```

```python
### Review of conditional selection 

df
```

```python
df[df['col1']>2]
```

```python
## Recall that df['col1']>1 is a boolean Series 

df['col1']>2
```

```python
### Using multiple conditions 

df[(df['col1']>2) & (df['col2']==444)]
```

```python
### THe apply method -> One of the most powerful methods in your Pandas toolbox 

##Consider the function

def times2(x):
    return x*2
```

```python
### Recall that we can call built-in functions from Data Frame columns as follows: 

df['col1'].sum()
```

```python
### Now, what if you want to apply your own custom function such as times2()?

##We can do so using the .apply() method passing the function we want to apply as an argument: 

df['col1'].apply(times2)
```

```python
###We can also pass built-in functions, e.g., obtaining the length of each String in column 3

df['col3'].apply(len)
```

```python
### The .apply() method canbe specially powerful when use it with lambda expressions -> No need to define the function

df['col2'].apply(lambda x:x**2) ##Square each value in column 2
```

```python
##Removing columns 

##Using the drop method 
df.drop('col1', axis = 1)
```

```python
##Dropping a row 
df.drop(0)
```

```python
###Obtaining column and index names of a Data Frame -> Atributes of the Data Frame itself 

df.columns 
```

```python
df.index ##We have a range index 
```

```python
###Sorting a Data Frame 
df.sort_values(by = 'col2') ##Pass in the column you want to sort by - Index is attached to the new sorted rows! 
```

```python
##Finding null NaN values in your Data Frame
df.isnull() ##Returns a boolean Data Frame indicating whether the values were null or not
```

```python
### Pivot table 

data = {'A':['foo','foo','foo','bar','bar','bar'],
     'B':['one','one','two','two','one','one'],
       'C':['x','y','x','y','x','y'],
       'D':[1,3,2,5,4,1]}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.pivot_table(values = 'D', index = ['A','B'], columns = ['C'] ) ##Takes three arguments: values, index and columns
```
