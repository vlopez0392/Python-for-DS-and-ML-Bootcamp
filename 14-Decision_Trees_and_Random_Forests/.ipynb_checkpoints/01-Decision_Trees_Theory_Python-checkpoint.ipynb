### Decision Trees and Random Forests Theory 

```python
### -----------------------------------------DECISION TREES------------------------------------------------------
### Decision Trees -> In a typical Decision Tree we have the following: 

### Nodes: Split for the value of a certain attribute (Years > 5) 
### Edges: Outcome of a split to the next node (Programming Languages -> Programming Languages > 3)

### How to choose the best split? -> Entropy and Information Gain are the mathematical methods of choosing the best
### split. -> In here we consider classification trees -> In classification trees, we predict that that each
### each observation belongs to the most commonly occurring class of the training observations in the region to 
### which it belongs.

### Intuition -> Try to choose the features that best split the data 


### -----------------------------------------RANDOM FORESTS ------------------------------------------------------

### Decision trees weakness -> Don't have the best predictive accuracy due to high variance.
###                         -> Different splits on the training data can lead to very different trees. 
### To reduce the variance we implement bagging.

### Bagging -> Random forests -> We create an ensemble of decision trees using bootstrap samples of the training 
                            ##   set. 
                            ##   Bootstrap samples -> Sampling from the training set with replacement. 
        ##                    -> We can use many trees with a random sample of m features chosen as the split.
        ##                    -> A new random sample of m features is chosen for every single tree at every single 
        ##                       split from a whole set of p features.
        ##                    -> For classification, m is typically chosen to be the square root of p        
        
### Whats the point? -> Suppose there is a very strong feature in the data set. When using "bagged" trees, most of
### the trees will use that feature as the top split, resulting in an ensemble of similar trees that are highly 
### correlated.  -> We want to avoid this because averaging highly correlated quantities DO NOT reduce variance. 

### Then, random forests decorrelates trees, such that the averaging process can reduce the variance of the 
### resulting model. 
```

### Decision Trees and Random Forests with Python

```python
### Imports 
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

##Reading the data 
df = pd.read_csv('kyphosis.csv')
```

```python
df.head() ## Kyphosis - Spinal Condition -> This dataset shows if the condition was present or not after the operation
          ## Age data in months - Children data
          ## Number - Number of vertebrae involved in the operation 
          ## Start  - Number of the first or topmost vertebrae that was operated on 
```

```python
### Exploring the data 

df.info() ## Very small data set
```

```python
sns.pairplot(df, hue = 'Kyphosis', diag_kind='hist')
```

```python
from sklearn.model_selection import train_test_split
```

```python
## Lets split our data into training and test sets 

X = df.drop('Kyphosis', axis = 1)
y = df['Kyphosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
## Lets import the Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
## Training the data
dtree.fit(X_train, y_train)
```

```python
## Lets do some predictions on our test set
predictions = dtree.predict(X_test)
```

```python
## Evaluate the model 
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
print('\n')
print(classification_report(y_test, predictions))
```

```python
### Lets see how this results compare to a random forest model 
from sklearn.ensemble import RandomForestClassifier ##RandomForest is an ensemble of many decision trees

rfc = RandomForestClassifier(n_estimators = 200)
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
## Evaluation of this model depens on whether the condition is absent vs. present
## Depends on the situation and the patient itself and the cost associated with each decision

print(confusion_matrix(y_test, rfc_pred))
print('\n')
print(classification_report(y_test, rfc_pred))

## Due to the size of the data set we obtain similar performanced for both algorithms (DT vs RF)
## For larger data sets, however, RF will outperform DT. 
```
