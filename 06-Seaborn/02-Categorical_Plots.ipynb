# 02 - Categorical Plots

```python
##Categorical plots -> Plotting categorical data 
import seaborn as sns
import numpy as np

%matplotlib inline 
tips = sns.load_dataset('tips')

tips.head()
```

```python
## For categorical data sets we are going to be concerned about seeing the distributions of a categorical column 
## such as sex, gender, day time, etc... 

## A - Barplots 
sns.barplot(x = 'sex', y ='total_bill', data= tips, palette = 'BuGn', estimator = np.std)
              ## A general plot that lets you aggregate the categorical data based off of some function 
              ## By default, that is the mean. -> Think of this as a visualization of a groupby action 
              
              ##We can change the estimator object to affect the type of grouping or aggregation presented. 
```

```python
## B - Countplot -> A count plot is similar to a barplot except that the estimator is set to count 
## the number of occurrences 

sns.countplot(x = 'sex', data = tips)
```

```python
## C - Boxplots and violin plots -> Are used to show the distribution of categorical data 
##     Boxplots also known as "box and whiskers" plots - SHows the distribution of quantitative data in a way that
##     hopefully facilitates comparisons between the variables 

sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker' ) ## Shows the quartiles of the distribution 
                                                                       ## Outside points are outliers
                                                                       ## Hue parameter - Allows to split the boxplots by 
                                                                       ## another categorical column  
            
            ## This is the true power of seaborn! We can extract useful data from our plots - Smokers tend to 
            ## have pay a smaller total bill on Fridays :) 
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True) 

## Similar to a boxplot.
## Allows us to plot all the components that correspond 
## to the data points. 
## Shows the KDE's of the underyling distributions.
## Shows the distribution of the points on its side.
## A little harder to grasp   

## Also provides the hue parameter, in addition, we can split the violin plot to visalize two distributions. e.g.
## Now we can directly compare two distributions -> Again, hard to visualize 

## Know your audience when using these plots! 
```

```python
## D - Strip plot 

ax = sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = True, hue = 'sex', dodge = True) 
                                                        ## Scatter plot based off a category
                                                        ## jitter parameter - Adds some random noise for better visualization 
                                                        ## Both the hue and split (dodge) parameters are available.
ax.legend(loc = 0)
```

```python
## E- Swarm Plot -> Combination of a violin and strip plot -? Points do not overlap and gives a better sense of the 
##                  distribution of values. -> Drawback -> DO NOT SCALE WELL for large datasets. Not used very often.    

sns.swarmplot(x = 'day', y = 'total_bill', data = tips)
```

```python
## We can combine violin and swarm plots, e.g. (Often used to explore data by Data Scientist themselves) 

sns.violinplot(x = 'day', y = 'total_bill', data = tips)
sns.swarmplot(x = 'day', y = 'total_bill', data = tips, color = 'black')
```

```python
## F- Cat Plot -> Is the most general form of all these plots - Using the kind parameter we can show any of the plots 
## shown above.
                    
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'violin') 
```
